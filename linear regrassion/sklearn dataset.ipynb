```python
# make datafream
import pandas as pd
df = pd.DataFrame(housing.data,
                   columns = housing.feature_names) # input data
df["MedHouseVal"] = housing['target']
df
```

```python
X = df.drop('MedHouseVal',axis=1)
y = df['MedHouseVal']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,  # Input data
                                                    y,  # output data
                                                    random_state=1234, # it select random samples
                                                    test_size=0.30)   #30 % test data hai means 70:30 ka ratio hai
```

```python
# from sklearn.linear_model import LinearRegression
# LR=LinearRegression()
# LR.fit(X_train,y_train)

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
# Model predictions happens X_test
y_predictions=model.predict(X_test)
y_predictions
```

```python
test_data=X_test
test_data['y_actual']=y_test
test_data['y_predictions']=y_predictions
test_data   
```

```python
print(y_test.values[:5])  # float 5. means 5.0
print(y_predictions[:5])
```

```python
# RMSE
# MSE
# MAE
# R-square

from sklearn.metrics import r2_score,mean_squared_error

R2=r2_score(y_test,y_predictions)

MSE=mean_squared_error(y_test,y_predictions)
#MSE**(1/2)

RMSE=np.sqrt(MSE)
#accuracy_score(y_test,y_predictions) # it is a regression tech

print("R-sqaure:",R2)
print("MSE:",MSE)
print("RMSE:",RMSE)
```
